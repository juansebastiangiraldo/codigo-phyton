```python
def determinar_num(numero):
  if numero > 0:
    return "el numero es positivo"
  elif numero < 0:
    return "el numero es negativo"
  elif numero == 0:
    return "el numero es 0"
```

```python
print(determinar_num(-2))
```

```python
def factorial(numero):
  if numero == 0:
    return 1
  else:
    return numero * factorial(numero - 1)
```

```python
print(factorial(6))
```

```python
def vocales_palabra(palabras):
  vocales = "aeiouAEIOU"
  contador = 0
  for letra in palabras:
    if letra in vocales:
      contador += 1
  return contador

```

```python
print(vocales_palabra("hola wordl"))
```

```python
def tabla_multiplicar(numero):
  for n in range(1, 11):
   print(f"{numero} * {n} = {numero * n}")
```
